```python
%matplotlib inline
import os
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image
import timeit
import cv2
import scipy
from scipy.ndimage import convolve
from scipy.ndimage import gaussian_laplace
from scipy.ndimage import rank_filter 
import skimage
from skimage.feature import peak_local_max, blob_dog, blob
```

```python
def show_all_circles(image, cx, cy, rad, color='r'):
    """
    image: numpy array, representing the grayscsale image
    cx, cy: numpy arrays or lists, centers of the detected blobs
    rad: numpy array or list, radius of the detected blobs
    """
    import matplotlib.pyplot as plt
    from matplotlib.patches import Circle

    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.imshow(image, cmap='gray')
    for x, y, r in zip(cx, cy, rad):
        circ = Circle((x, y), r, color=color, fill=False)
        ax.add_patch(circ)

    plt.title('%i circles' % len(cx))
    plt.show()
```

```python
def read_image(IMG_NAME):
    # YOUR CODE HERE
    gray=np.asarray(Image.open(IMG_NAME).convert('L'))/255
    return gray
```

```python
def filter_image_iteratiion(IMG,sigma_initial, n, multi_factor):
    
    h, w = IMG.shape

    scale_space = np.empty((h,w,n)) # [h,w] - dimensions of image, n - number of levels in scale space
    scale_fact = 1
    r = []
    for i in range(n):
        sigma=sigma_initial*multi_factor**i
        r.append(np.sqrt(2)*sigma)
        scale_space[:,:,i] = sigma**2*gaussian_laplace(IMG,sigma)

    return scale_space, r 
```

```python
def nonmaximum_suppression(img_stack):
    
    print(img_stack.shape)
    non_max_return = np.zeros([img_stack.shape[0],img_stack.shape[1],img_stack.shape[2]])
    peak_index= peak_local_max(scale_space)
    non_max_return[peak_index[:,0],peak_index[:,1],peak_index[:,2]] = img_stack[peak_index[:,0],peak_index[:,1],peak_index[:,2]]
    thred = np.where(non_max_return>0.015,non_max_return,0)
    cxy = np.nonzero(thred)
    
    
    return cxy
  
```

```python
IMG_FOLDER = 'part2_images/'
IMG_NAME = 'butterfly.jpg'
initial_sigma = 2
num_iter = 10
factor = 1.5
img = read_image(IMG_FOLDER + IMG_NAME)
scale_space ,r  = filter_image_iteratiion(img,initial_sigma,num_iter,factor)

cxy= nonmaximum_suppression(scale_space)
cx = cxy[1]
cy = cxy[0]
cz = cxy[2]
r = np.array(r)
cr = r[cz]

show_all_circles(img, cx, cy, cr)
```
